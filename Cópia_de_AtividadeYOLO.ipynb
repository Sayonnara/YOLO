# Vendo a Imagem

---



```python
imagem = cv2.imread("1.jpg")
cv2_imshow(imagem)

altura, largura, profundidade = imagem.shape
pixels = imagem.size
print("NY - imagem de teste:\n",
      "\tAltura:", altura, "pixels;\n"
      "\tLargura:", largura, "pixels;\n"
      "\tProfundidade:", profundidade, "canais;\n"
      "\tTotal de pixels:", pixels, '.')
```

# **Mudando o tamanho da imagem**

---



```python
redim = cv2.resize(imagem, (int(largura * 0.5), int(altura * 0.5)))
cv2_imshow(redim)
```

```python
imagem_2 =cv2.cvtColor(imagem, cv2.COLOR_RGB2BGR)
imagem_concatenada = cv2.hconcat((imagem, imagem_2))
cv2_imshow(imagem_concatenada)
```
